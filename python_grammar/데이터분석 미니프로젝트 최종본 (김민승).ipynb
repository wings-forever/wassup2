#### 2) 최근 학습 성취도 (recent_learning_achievement)

target1의 사용자 수 6199명, target0의 사용자 수 3801명으로 두 집단의 최근 학습 성취도 평균을 비교하기엔 모집단의 크기 차이가 존재하기 때문에 표준화를 통해 두 집단의 평균 최근 학습 성취도를 비교할 예정

- 참고자료: # 표준화 vs 정규화 - zscore, minmax scaler|작성자 키아     
    https://blog.naver.com/kiakass/222028896737

```python
from sklearn.preprocessing import minmax_scale
import numpy as np 

temp_target1 = minmax_scale(target1["recent_learning_achievement"])
temp_target0 = minmax_scale(target0["recent_learning_achievement"])

temp_target1 = round(np.mean(temp_target1),2)
temp_target0 = round(np.mean(temp_target0),2)

print("target1 표준화 :",temp_target1)
print("target0 표준화 :",temp_target0)
```

```python
plt.figure(figsize=(10,5))

plt.bar("target1의 평균 최근 학습 성취도",temp_target1)
    
plt.bar("target0의 평균 최근 학습 성취도",temp_target1)

plt.show()
```

- 다음 달 구독을 연장할 사용자 (target 1)과 다음 달 구독을 취소할 사용자 (target 0)의 최근 학습 성취도 평균값을 비교해 보면, 그래프 상으로 유의한 차이가 보이지 않는다.     
- 수치상으로는 target1의 경우 0.51, target0의 경우 0.55로 target0 즉, 다음 달 구독을 취소할 사용자의 최근 학습 성취도가 0.04정도 더 높은 것을 확인할 수 있다. 따라서 "최근 학습 성취도" 데이터는 구독을 연장할 사용자 예측에 무관한 지표라 결론지을 수 있다.

따라서 월간 활동적인 학습 일수가 1일에서 10일의 구간에 해당하는 사용자는 구독을 연장할 가능성이 낮을 수 있다고 예측할 수 있으며, 학습 일수가 21일 이상인 사용자는 구독을 연장할 가능성이 높은 사용자로 예측할 수 있다.
